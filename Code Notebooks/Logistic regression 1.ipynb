```python
y1 = data.drop('Cabin',axis=1)
```

```python
y1
```

```python
y1 = y1.dropna()
```

```python
sns.heatmap(y1.isnull())
```

```python
y1.isnull().sum()
```

```python
y1
```

```python
#pd.get_dummies(y1.Sex)
#pd.get_dummies(y1['Sex'])
gendery1 =pd.get_dummies(y1['Sex'],drop_first=True)
gendery1
```

```python
#embarkedy1 = pd.get_dummies(y1['Embarked'])
embarkedy1 = pd.get_dummies(y1['Embarked'],drop_first=True)
embarkedy1
```

```python
#pclassy1 = pd.get_dummies(y1['Pclass'])
pclassy1 = pd.get_dummies(y1['Pclass'],drop_first=True)
pclassy1
```

```python
y2 = pd.concat([y1,gendery1,embarkedy1,pclassy1],axis=1)
```

```python
y2
```

```python
y2=y2.drop(['PassengerId','Sex','Embarked','Name','Ticket','Pclass'],axis=1)
```

```python
y2
```

```python
Y = y2.Survived
Y
```

```python
X = y2.drop('Survived',axis=1)
X
```

```python
from sklearn.model_selection import train_test_split
#from sklearn.cross_validation import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,random_state=100,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
Ypred = model.predict(x_test)
```

```python
Ypred
```

```python
from sklearn.metrics import classification_report
```

```python
classification_report(y_test,Ypred)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test,Ypred)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,Ypred)
```
